## Compare the PCA and Least Squares Fitting Method

- Run the PCA_Linear/PCA_Linear.py to get the output
- Compare the RMSE, Angular Deviation and Initial point

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Run PCA_Linear/PCA_Linear.py to get the output
import PCA_Linear.PCA_Linear
```

```python
particle_info = pd.read_csv('PCA_Linear/particle_info.csv')
```

```python
# Compare the RMSE
PCA_RMSE = particle_info['PCA_RMSE'].values.T
Linear_RMSE = particle_info['Linear_RMSE'].values.T
plt.plot(PCA_RMSE, label='PCA RMSE', color='blue', marker='o')
plt.plot(Linear_RMSE, label='Linear RMSE', color='red', marker='x')
plt.title('RMSE Comparison')
plt.legend()
plt.show()
```

```python
# Compare the Angular Deviation
PCA_angle = particle_info['PCA_angle'].values.T
Linear_angle = particle_info['Linear_angle'].values.T
plt.plot(PCA_angle, label='PCA angle', color='blue', marker='o')
plt.plot(Linear_angle, label='Linear angle', color='red', marker='x')
plt.title('Angular Deviation Comparison')
plt.legend()
plt.show()
```

```python
# Compare the Initial Point
PCA = particle_info[['PCA_init_x', 'PCA_init_y']].values
Linear = particle_info[['Linear_init_x', 'Linear_init_y']].values

plt.scatter(PCA[:, 0], PCA[:, 1], label='PCA', color='blue', marker='o')
plt.scatter(Linear[:, 0], Linear[:, 1], label='Linear', color='red', marker='x')
plt.title('Initial Point Comparison')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
